```python
import os, shutil
```

```python
INPUTFOLDER = '/home/haehn/Dropbox/RESEARCH/MITOCHONDRIA/haehnassignment1_align/haehnassignment1_data/'
```

```python
OUTPUTFOLDER = '/home/haehn/Dropbox/RESEARCH/MITOCHONDRIA/xp1/'
```

```python
# source images
files = os.listdir(INPUTFOLDER+'/Train_Images_Star')
for f in files:
    print(f)
    new_name = f.replace('.png', '_source.png')
    shutil.copyfile(INPUTFOLDER+'/Train_Images_Star/'+f, OUTPUTFOLDER+'/train/'+new_name)
```

```python
# target images
files = os.listdir(INPUTFOLDER+'/Train_Images')
for f in files:
    print(f)
    new_name = f.replace('.png', '_target.png')
    shutil.copyfile(INPUTFOLDER+'/Train_Images/'+f, OUTPUTFOLDER+'/train/'+new_name)
```
